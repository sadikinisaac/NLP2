```
import matplotlib.pyplot as plt
import nltk
from nltk.probability import FreqDist
```

```
words = 'I like you you and and you like me good good good'.split()
```

```
fd_words = FreqDist(words)
plt.figure(figsize=(20, 5))
fd_words.plot(10)
```
